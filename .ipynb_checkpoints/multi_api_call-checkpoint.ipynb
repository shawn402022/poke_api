```python
def multi_poke_api_call(a_list=None):
    if not a_list:
        while True:
            pokemon = input("name your pokemon: ")
            response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon}')
            

            if response.status_code == 200:
                print('success')
                

            else:
                print(f'error,status code {response.status_code}')
                print('check spelling')
                
                
            addFinDel = input("press 'a' to add or press, 'p' to print pokemon info, 'd' to delete last entry: ")
            
            
                        
            if addFinDel == 'a':
                your_team.append(pokemon)
                print(f'success, {pokemon} has been added to your_team')
                print(your_team)
            elif addFinDel == 'd':
                your_team.append(pokemon)
                dP = your_team[-1]
                print(f'{dP} has been deleted from your_team')
                del your_team[-1]
                
                length = len(your_team)
                if length == 0:
                    print('your current team is empty')
                else:
                    print(f'your current team is {your_team}')
                
            elif addFinDel == 'p':
                your_team.append(pokemon)
                print(f'success, {pokemon} has been added to your_team')
                print(f'your pokemon team consisist of {your_team} ')
                
                
                #this part of the code prepares and declares dictionaries 
                
                data = response.json()

                pAbility = [item['ability']['name'] for item in data['abilities']]



                name = data['name']
                types = data['types'][0]['type']['name']
                weight= data['weight']
                abilities = pAbility
                height = data['height']
                front = data['sprites']['front_default']

                pokemon_dict = {
                    'name':name,
                    'types' : types,
                    'weight' : weight,
                    'ability' : abilities,
                    'height' : height,
                    'front' : front,
                    
                }
                
                for pokemon in your_team:
                    print(pokemon_dict)


                
    



            
            

            
        
multi_poke_api_call()

            
            
        
            
            
    


```
